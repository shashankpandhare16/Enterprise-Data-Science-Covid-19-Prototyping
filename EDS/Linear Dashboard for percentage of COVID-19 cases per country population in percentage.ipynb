#### Plot.ly

```python
import dash
dash.__version__
from dash import dcc
from dash import html

app = dash.Dash()
app.layout = html.Div([
    
    html.Label('Multi-Select Country'),
    
       dcc.Dropdown(
        id='country_drop_down',
        options=[{'label': i, 'value': i}
                for i in df_plot_all['location'].unique() ],
           
        value=['India'], # which are pre-selected
        multi=True
    ),   
        
    dcc.Graph(id='main_window_slope', figure={})
])
```

```python
from dash.dependencies import Input, Output

@app.callback(
    Output('main_window_slope', 'figure'),
    [Input('country_drop_down', 'value')])
def update_graph(option):
    print(option)
    print(type(option))
    
    if type(option)!=str:
        dff = df_plot_all[df_plot_all['location'].isin(option) ]
    else:
        dff = df_plot_all[df_plot_all['location']==option]
        

    
    fig = px.line(dff, x='date', y='relative_cases_in_%',color='location',title='Covid Relative Cases Over Time in linear scale')
    
#     fig = px.line(df_my_list, x="date", y='total_cases_ratio_in_%', color ='location', title='COVID cases')
#     fig.update_layout(yaxis_type="log")
    
    return fig
```

```python
app.run_server(debug=True, use_reloader=False)
```
