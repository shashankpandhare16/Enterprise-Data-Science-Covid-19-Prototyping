### Linear Graph People vaccinated over time for five countries

```python
fig = px.line(df_my_list, x="date", y="people_vaccinated", color ='location', title='People Vaccinated')
fig.show()
```

```python
df_plot1 = df_my_list[(df_my_list['date']>'2020-11-01')]
```

```python
fig = px.line(df_plot1, x="date", y="people_vaccinated", color ='location', title='Linear Graph of People Vaccinated')
fig.show()
```

```python
df_my_list['people_vaccination_ratio_in_%'] = (df_my_list['people_vaccinated']/df_my_list['population'])*100
```

###  Figure 2 The vaccination rate (percentage of the population) over time in percentage

```python
fig = px.line(df_plot1, x="date", y='people_vaccination_ratio_in_%', color ='location', title='Vaccination rate in %')
fig.show()
```

## Logarithmic Graph : The vaccination rate (percentage of the population) over time in percentage

```python
fig = px.line(df_plot1, x="date", y='people_vaccination_ratio_in_%', color ='location', title='Logarithmic Graph of Vaccination rate w.r.t. population in percentage')
fig.update_layout(yaxis_type="log")
fig.show()
```

# Plotting the people vaccinated w.r.t. population  in percentage on a World Map

```python
data = dict(type='choropleth',
            colorscale = 'Viridis',
            reversescale = True,
           locations = df_my_list['location'],
            locationmode = "country names",
           z = df_my_list['people_vaccination_ratio_in_%'],
           text = df_my_list['location'],
           colorbar = {'title' : 'Vaccination rate of countries with respect to population in percentage'},
           
           
           )
```

```python

layout = dict(title = 'COVID 19 Vaccination rate distribution of total cases w.r.t population in %',
                geo = dict(showframe = False,projection = {'type':'mercator'})
             )
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```
